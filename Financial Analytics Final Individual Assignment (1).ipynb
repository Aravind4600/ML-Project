# Question 4

A. What’s the relationship between decision trees and Random Forest? What are some of the pros/cons of each approach?

The relationship between the Decision Tree and Random Forest would be that Decision Tree uses a tree-like model of decisions and their possible consequences. It's a simple yet powerful tool for both classification and regression tasks while Random Forrest uses multiple decision trees to make the better predictions. The key relationship between the two are 

1.Random Forrest is a collection of the decision trees where each tree is independent of others and final prediction is a function of the prediction of the trees 

2.Training for the decision trees are done on entire dataset while in Random Forrest, each tree is trained for different bootstrap sample of the data. Random Forrest employes feature randomness where each tree is allowed to consider a subset of feature at each split.

3.Predictions made by a decision tree is following its learned decision paths, while a Random Forrest makes its predictions by aggregating the prediction of all the decision trees.

The Pros and Cons Decision Trees would be
Pros 
Easy to interpret and visualize.

It is able handle both numeric and categorical data. 

Doesn’t require a lot of preprocessing, scaling or normalization. 

Can model nonlinear relationships. 

Cons 

Prone to overfitting if the tree is deep. 

Sensitive to small changes in data which can result in different tree. 

Decision trees are high variance classifiers, where data is split into two parts randomly, the resulting decision trees can be quite different. 

The Pros and Cons Random Forrest would be

Pros 

Provide more accurate predictions than individual decision trees.

Less likely to overfit as they average the results of many trees. 

They can handle large datasets. 

Cons

More complex and computationally expensive due to large number of trees 

Less interpretable than individual decision trees 

B. List three techniques and how they are used to address imbalanced datasets in machine learning.

Imbalanced datasets are the datasets that have one class significantly outnumbers the other which can can cause problems to the machine learning models .

The three techniques that can be done use to address he issues would be 

Resampling : It involves changing the dataset to have balanced datasets there are two methods of resampling which would be Oversampling (randomly adding more instances of minority class ) and under sampling (reducing instances of majority class)


SMOTE (Synthetic Minority Oversampling Technique ) This technique is type of oversampling  where instead of adding more instances by duplicating it synthesis new instances in the datasets rather than oversampling through random oversampling method 

Ensemble Methods : This involves building multiple models (usually decision trees) and combining them in a way that makes the overall model more robust and less prone to errors. For example, in "Balanced Random Forests", the algorithm applies under sampling to the majority class in each bootstrap sample before building each tree

# QUESTION 8

Based on the 2 stocks chosen in question 7 use this data to construct a Monte-Carlo simulation. What is the average predicted price for both stocks. Based on the results from both simulations comment on how this may affect your trading strategy.

#### Monte Calo Simulation for Amazon

```python
ticker = 'AMZN'
df2 = yf.download(ticker, period='10y')
data1 = df2[['Adj Close']]
data1.tail()
```

```python
returns = np.log(data1/data1.shift(1))
returns.head()
```

```python
data1.plot(figsize=(10,6))
```

```python
returns.plot(figsize=(10,6))
```

```python
rm = returns.mean()
```

```python
rm
```

```python
rvar = returns.var()
```

```python
rvar
```

```python
drift=rm-(.5*rvar)
```

```python
drift
```

```python
stdev = returns.std()
```

```python
stdev
```

```python
t_intervals = 250
iterations = 50
```

```python
x = np.random.rand(t_intervals, iterations)
x
```

```python
Z = norm.ppf(x)
Z
```

```python
daily_returns = np.exp(drift.values + stdev.values * Z)
```

```python
daily_returns
```

```python
drift.values
```

```python
type(drift)
```

```python
S0 = data1.iloc[-1]
S0
```

```python
price_list = np.zeros_like(daily_returns)
```

```python
price_list
```

```python
price_list[0] = S0
price_list
```

```python
for t in range(1, t_intervals):
    price_list[t] = price_list[t - 1] * daily_returns[t]
```

```python
price_list
```

```python
plt.figure(figsize=(10,7))
plt.plot(price_list);
```

```python
price_list.mean()
```

```python
price_list.max()
```

```python
price_list.min()
```

```python
np.percentile(price_list,5)
```

```python
np.percentile(price_list,95)
```
