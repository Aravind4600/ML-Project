```python
def time_to_days(time_str):
    parts = time_str.split()
    # Replace 'a' or 'an' with '1' if it's the first part of the string
    if parts[0] in ('a', 'an'):
        parts[0] = '1'
    try:
        num = int(parts[0])
    except ValueError:
        return 0
    unit = parts[1].lower()
    # Convert the time unit to days
    if 'day' in unit:
        return num
    elif 'week' in unit:
        return num * 7
    elif 'month' in unit:
        return num * 30  # Approximation, as months have varying number of days
    elif 'year' in unit:
        return num * 365  # Approximation, ignoring leap years
    else:
        return 0
```

```python
review_time_in_days = data['review_time'].apply(time_to_days)
```
